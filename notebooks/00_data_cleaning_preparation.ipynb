```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns   
from datetime import datetime
```

```python
df = pd.read_csv("../data/superstore_data.csv")
df.head()
```

```python
df.info()
```

```python
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], errors="coerce")
```

```python
df.describe()
```

```python
df = df[df["Year_Birth"] > 1920]
```

```python
missing = df.isnull().sum()
print("Missing values in each column:")
print(missing[missing > 0])
```

```python
df = df[df["Income"] < 200000] 
```

```python
df = df.dropna(subset=["Income"])
```

```python
df["Age"] = 2014 - df["Year_Birth"]
df["TotalChildren"] = df["Kidhome"] + df["Teenhome"]
df["Customer_Since_Years"] = 2014 - pd.to_datetime(df["Dt_Customer"]).dt.year
df["HasChildren"] = df["TotalChildren"].apply(lambda x: "Yes" if x > 0 else "No")
```

```python
df.describe()
```

```python
invalid_status = ["Alone", "YOLO", "Absurd"]
df = df[~df["Marital_Status"].isin(invalid_status)]
df["Marital_Status"].value_counts()
```

```python
education_map = {
    "Basic": "Undergraduate",
    "2n Cycle": "Undergraduate",
    "Graduation": "Graduate",
    "Master": "Postgraduate",
    "PhD": "Postgraduate"
}
df["Education"] = df["Education"].map(education_map)
df["Education"].value_counts()
```

```python
df.to_csv("../data/cleaned_superstore.csv", index=False)
```
